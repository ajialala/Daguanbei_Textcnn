```python
import pandas as pd
from collections import Counter
```

```python
file = pd.read_csv('cnn_data_letter',sep='\t')
```

```python
file.head()
```

```python
Counter(file['class'])
```

```python
file = file.sample(frac=1).reset_index(drop=True)
```

```python
file.head()
```

```python
fenge = [0.8,0.1,0.1]
```

```python
file.iloc[0:int(len(file)*fenge[0]),].to_csv('train_data_letter',index=0,header=0,sep='\t')
file.iloc[int(len(file)*fenge[0]):int(len(file)*(fenge[0]+fenge[1])),].to_csv('val_data_letter',index=0,header=0,sep='\t')
file.iloc[int(len(file)*(fenge[0]+fenge[1])):len(file),].to_csv('test_data_letter',index=0,header=0,sep='\t')
```
